```python
from itertools import product

maxRow, maxCol = (10, 10)

noiseList = np.array([[d1,d2,d3,d4] for d1 in range(7, 14)
                for d2 in range(7, 14)
                    for d3 in range(7, 14)
                        for d4 in range(7, 14)])*.1

maxStep=11

move = np.array([[1,0],[0,1],[0,-1],[-1,0]])

path = np.intp([np.ones((maxRow, maxCol))*-1]*(maxStep+1))

prob = np.float64([np.ones((maxRow, maxCol))]*(maxStep+1))

for step in range(maxStep):
    
    for cell in product(range(maxRow), range(maxCol)):
        row, col = cell
        if grid[row,col]!=1:
            continue
        
        if step>0:
            if row==0: 
                select=3
            else:
                select=0
            selectedStep = cell - move[select]
            for option in range(4):
                lastStep=cell-move[option]
                r, c = lastStep
                if r>-1 and r<maxRow and c>-1 and c<maxCol and grid[r,c]==1:
                        if prob[step-1][r][c]<prob[step][selectedStep[0]][selectedStep[1]]:
                            select=option
                            selectedStep = cell - move[select]
            path[step][row][col]=select
        
        dist = np.sqrt(np.sum(np.square(target - cell), axis=1))

        for noise in noiseList:
            test = np.round(np.multiply(dist, noise)*10.0)/10.0
            if np.sum(np.square(test-noisyDist[step,:]))<prob[step][row,col]:
                prob[step][row,col]=np.sum(np.square(test-noisyDist[step,:]))
                
        if step>0:
            select = path[step][row][col]
            selectedStep = cell - move[select]
            prob[step][row,col]*=prob[step][selectedStep[0]][selectedStep[1]]
            
    print 'Step ',step, prob[step]
```

```python
row = np.argmin(prob[maxStep-1])/maxCol
col = np.argmin(prob[maxStep-1])%maxCol

solution=[]
for step in reversed(range(maxStep)):
    solution.append((row,col))
    print row,col,path[step][row,col]
    row,col=np.subtract([row,col],move[path[step][row,col]])
    
solution=[(r,c) for r,c in reversed(solution)]
solution
```
