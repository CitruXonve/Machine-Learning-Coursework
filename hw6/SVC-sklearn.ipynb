```python
import numpy as np

f = open('linsep.txt','r')
data=f.readlines()
data=np.array(map(lambda x:map(eval,x[:-1].split(',')),data))
x, y=data[:,:2].tolist(),data[:,2:].tolist()
x,y
```

```python
# import support vector classifier 
from sklearn.svm import SVC # "Support Vector Classifier" 
clf = SVC(kernel='linear') 
  
# fitting x samples and y classes 
clf = clf.fit(x, y)
clf
```

```python
import matplotlib.pyplot as plt

xfit = np.linspace(0.0, 1.0) 
# plt.plot(xfit,clf.coef_*xfit+clf.intercept_,'-k')
plt.scatter(np.array(x)[:,0].tolist(),np.array(x)[:,1].tolist(), s=50, cmap='spring')
# plt.plot(xfit, -xfit*clf.coef_[0,1]/clf.coef_[0,0], '-k')

def plot_separator(ax, w, b):
    slope = -w[0] / w[1]
    intercept = -b / w[1]
    x = np.arange(0, 2)
    ax.plot(x, x * slope + intercept, 'k-')

plt.show() 

clf.coef_, clf.intercept_
```
