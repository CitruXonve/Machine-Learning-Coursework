```python
import numpy as np

f = open('linsep.txt','r')
data=f.readlines()
data=np.array(map(lambda x:map(eval,x[:-1].split(',')),data))
x, y=data[:,:2].tolist(),data[:,2:].tolist()
x,y
```

```python
# import support vector classifier 
from sklearn.svm import SVC # "Support Vector Classifier" 
clf = SVC(kernel='linear') 
  
# fitting x samples and y classes 
clf = clf.fit(np.array(x), y)
clf
```

```python
import matplotlib.pyplot as plt

xfit = np.linspace(0.0, 1.0) 
# plt.plot(xfit,clf.coef_*xfit+clf.intercept_,'-k')
plt.scatter(np.array(x)[:,0].tolist(),np.array(x)[:,1].tolist(), c=np.array(y).flatten(), s=50, cmap='spring')
# plt.plot(xfit, -xfit*clf.coef_[0,1]/clf.coef_[0,0], '-k')

def plot_separator(plt, w, b):
    slope = -w[0] / w[1]
    intercept = -b / w[1]
    x = np.linspace(0, 1.0)
    plt.plot(x, x * slope + intercept, '-k')

w, b = clf.coef_.flatten(), clf.intercept_[0]
    
plot_separator(plt,w,b)
plt.show() 

clf.coef_.tolist(), clf.intercept_.tolist()
```
